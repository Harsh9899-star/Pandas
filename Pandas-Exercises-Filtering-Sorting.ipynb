# Filtering and Sorting Data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset *chipotle.tsv*

```python
chipo = pd.read_table("Exercise Data Files/Pandas Exercise Filtering Sorting Files/chipotle.tsv",sep="\t")
```

```python
chipo
```

### Step 3. Assign it to a variable called chipo.

```python
chipo
```

```python
chipo.info()
```

```python
chipo['item_price']=chipo['item_price'].str.replace('$','')
```

```python
chipo['item_price']=chipo['item_price'].astype(float)
```

### Step 4. How many products cost more than $10.00? 

Use `str` attribute to remove the $ sign and convert the column to proper numeric type data before filtering.


```python
chipo.iloc[np.where(chipo['item_price']>10)]
```

### Step 5. What is the price of each item? 

print a data frame with only two columns `item_name` and `item_price`

```python
chipo.groupby(['item_name'])['item_price'].unique()
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by='item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
x=chipo.iloc[np.where(chipo['item_price'])].max()
print(x)
x.quantity
```

```python
chipo.groupby(['item_name'])['quantity'].agg(['max'])
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
a=chipo.iloc[np.where(chipo['item_name']=='Veggie Salad Bowl')]
a.count()
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
b=chipo.iloc[np.where(chipo['item_name']=="Canned Soda")]
c=b.quantity>1
c.count()
```
